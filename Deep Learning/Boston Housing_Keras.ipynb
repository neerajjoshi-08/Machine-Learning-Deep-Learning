### Boston Hosing Dataset using keras

```python
# Importing library
import tensorflow as tf
```

##### 1.Collecting dataset

```python
(data, target),_ = tf.keras.datasets.boston_housing.load_data(test_split=0)
```

```python
data[:5, :]
```

```python
target[:5]
```

##### 2. Building a graph

```python
# Intialising sequential graph.
model = tf.keras.models.Sequential()

# Normalising the input data.
model.add(tf.keras.layers.BatchNormalization(input_shape=(13,)))

# Adding Dense layer (y= Wx + b)
model.add(tf.keras.layers.Dense(1))   # NO need to provide the shape as it's sequential model.
```

```python
# Compiling the model - GradientDescent and loss (MSE).
model.compile(optimizer='sgd', loss='mse')   # Default learning rate = 0.01
```

##### 3. Executing the graph

```python
model.fit(data, target, epochs=100)
```

```python
# Changing the value of learning rate to 0.03
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.03), loss='mse')   
model.fit(data, target, epochs=100)
```
