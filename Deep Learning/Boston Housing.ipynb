### Boston Housing Dataset using tensorflow

```python
# Importing libraries
import tensorflow as tf
import numpy as np
from tensorflow.contrib.learn import datasets
```

```python
# Resetting Default graph so as to refrain from duplicate nodes [Only need in Jupyter Notebook]
tf.reset_default_graph()
```

##### 1. Loading boston dataset

```python
boston = datasets.load_dataset('boston')
```

```python
# Input features
data = np.array(boston.data)

# Acutal price
target = np.array(boston.target)
```

```python
data
```

```python
n_features = data.shape[1]
```

##### 2. Buliding a graph or model

```python
# 2.1 Defining Input features and Actual output.
# Input
x = tf.placeholder(shape=[None, n_features], dtype=tf.float32, name='x-input')

# Output
y_ = tf.placeholder(shape=[None], dtype=tf.float32, name='y-input')
```

```python
# Normalizing the data
x_n = tf.layers.batch_normalization(x, training=True)
```

```python
# 2.2 Define the weights and bias.
W = tf.Variable(tf.zeros(shape=[n_features, 1]), name='weights')
b = tf.Variable(tf.zeros(shape=[1]), name='bias')
```

```python
# 2.3 Prediction
y = tf.add(tf.matmul(x_n, W), b, name='output')
```

```python
# 2.4 Loss
loss = tf.reduce_mean(tf.square(y-y_), name='loss')   # Mean Square Error
```

```python
# 2.5 Defining Gradient Descent Optimiser
train_op = tf.train.GradientDescentOptimizer(0.03).minimize(loss)
```

##### 3. Executing the graph or model

```python
sess = tf.Session()
```

```python
# 3.1 Initilasing the variables. 
sess.run(tf.global_variables_initializer())
```

```python
# 3.2 Executing node
trainingEpoch = 1000
for epoch in range(trainingEpoch):
    # Calculating train_op and loss
    _, trainloss = sess.run([train_op, loss], feed_dict={x: data, y_: target})
    
    if epoch % 100 == 0:
        print(f'Training loss at #{epoch} is: {trainloss}')
```

```python
sess.close()
```
