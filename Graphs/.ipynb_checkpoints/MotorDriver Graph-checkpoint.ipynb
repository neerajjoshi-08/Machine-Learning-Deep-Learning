```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
motor = pd.read_csv('motor.csv')
```

```python
motor.head()
```

```python
plt.title("Supply Voltage Vs Efficient (%)")
sns.pairplot(motor['Supply Voltage (V)'], motor['Efficient (%)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Supply Voltage Vs Revolving Speed (RPM)")
sns.barplot(motor['Supply Voltage (V)'], motor['Revolving Speed (RPM)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Efficient (%)")
sns.distplot(motor['Efficient (%)'], kde=False, rug=True);
```

```python
plt.title("Revolving Speed (RPM)")
sns.distplot(motor['Revolving Speed (RPM)'], kde=False, rug=True);
```

```python
plt.title("Supply Voltage (V) Vs Revolving Speed (RPM)")
sns.stripplot(motor['Supply Voltage (V)'], motor['Revolving Speed (RPM)'], hue = motor['The DC motor Driver'], jitter = True)
```

```python
plt.title("Supply Voltage (V) Vs Efficient (%)'")
sns.stripplot(motor['Supply Voltage (V)'], motor['Efficient (%)'], hue = motor['The DC motor Driver'], jitter = True)
```

```python
plt.title("Supply Voltage (V) Vs Efficient (%)'")
sns.boxplot(motor['Supply Voltage (V)'], motor['Efficient (%)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Supply Voltage (V) Vs Efficient (%)")
sns.boxplot(motor['Supply Voltage (V)'], motor['Efficient (%)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Supply Voltage (V) Vs Revolving Speed (RPM)")
sns.barplot(motor['Supply Voltage (V)'], motor['Revolving Speed (RPM)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Supply Voltage (V) Vs Revolving Speed (RPM)")
sns.pointplot(motor['Supply Voltage (V)'], motor['Revolving Speed (RPM)'], hue = motor['The DC motor Driver'])
```

```python
plt.title("Supply Voltage (V) Vs Efficient (%)")
sns.pointplot(motor['Supply Voltage (V)'], motor['Efficient (%)'], hue = motor['The DC motor Driver'])
```

```python

sns.lmplot(x="Supply Voltage (V)", y="Efficient (%)", data=motor, hue = "The DC motor Driver")
plt.title("Supply Voltage (V) Vs Efficient (%)")
```

```python
sns.lmplot(x="Supply Voltage (V)", y="Revolving Speed (RPM)", data=motor, hue = "The DC motor Driver")
plt.title("Supply Voltage (V) Vs Revolving Speed (RPM)")
```
