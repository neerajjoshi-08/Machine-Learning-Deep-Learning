## Motor graphs

```python
df = pd.read_excel('Motor driver Table.xlsx')
motors = df['The DC motor Driver'].unique()
```

```python
# Row and column wise images.
fig, ax = plt.subplots(5, 3, figsize=(0.2,0.2)) #figsize=(3,1)
color = ['red', 'blue', 'green']
index = 0
for j in motors:
    df1 = df[df['The DC motor Driver'] == j]
    for i in range(1, 4):
        plt.subplots_adjust(right=60)
        sns.barplot(df1.iloc[:, 0], df1.iloc[:, i], ax=ax[index][i-1], color=color[i-1], ci=None)
        fig.show()
    index += 1
    plt.subplots_adjust( top=70)
    
    
# NOTE: Bar graph width is too large. Try to reduced it.
```
